```python

# Function to label comments using OpenAI API
def label_comment(comment):
    try:
        response = openai.ChatCompletion.create(
            model="gpt-3.5-turbo",
            messages=[
                {"role": "user", "content": f"Classify this comment as either 'hate' or 'not hate'. Analyze carefully for any hateful, offensive, or abusive language. The comment may be in multiple languages, may contain emojis, and may have mixed scripts: {comment}"}
            ]
        )
        return response['choices'][0]['message']['content'].strip()
    except Exception as e:
        print(f"Error processing comment: {comment}, Error: {str(e)}")
        return "error"

# check the valid comments
def is_valid_comment(comment):
    if not isinstance(comment, str):  # comment is a string
        return False
    return bool(re.search(r'[\w\s]|[\U0001F600-\U0001F64F]', comment, flags=re.UNICODE))

# Filter DataFrame to only valid comments
df = df[df['text'].apply(is_valid_comment)].copy()  # Avoid SettingWithCopyWarning

# Initialize a column for labels
df['label'] = ""  # Prepare the column for label assignment

# Set batch size and iterate over the entire dataset
batch_size = 10  # Adjust batch size based on rate limits and performance

for i in range(0, len(df), batch_size):
    batch = df['text'].iloc[i:i + batch_size]  # Slice using iloc
    labels = [label_comment(comment) for comment in batch]

    # Assign labels to the DataFrame slice
    df.iloc[i:i + len(labels), df.columns.get_loc('label')] = labels

    print(f"Processed batch {i} to {i + len(labels) - 1}")
    time.sleep(1)  # Adjust sleep time to avoid rate limits

# Save the labeled DataFrame to an Excel file
df.to_excel('OpenAi_labeled_comments_complete.xlsx', index=False)

# Display a quick preview of the labeled data
print(df[['text', 'label']].head())
```
